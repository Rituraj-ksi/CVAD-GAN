```python
import numpy as np
from sklearn.metrics import precision_recall_curve
import os
import matplotlib.pyplot as plt
```

```python
dir = "./tprfpr/"
folders = [x[0] for x in os.walk(dir) if x[0].endswith("avenue")]

print(folders)

recalls = []
precisions = []
for f1 in folders:
  precisionf = [os.path.join(f1,x) for x in os.listdir(f1) if x.startswith('precision')]
  recallf = [os.path.join(f1,x) for x in os.listdir(f1) if x.startswith('recall')]

  precision = np.load(precisionf[0])
  recall = np.load(recallf[0])

  precisions.append(precision)
  recalls.append(recall)
```

```python
# precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
plt.plot(recalls[0], precisions[0], '-.r')

plt.ylabel("Precision")
plt.xlabel("Recall")
# plt.title("Precision-Recall curve")

plt.legend(['Avenue']) 

plt.savefig('ped_aupr.pdf')
```
